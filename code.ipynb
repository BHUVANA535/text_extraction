```python
import gradio as gr
import pytesseract
from PIL import Image
import cv2
import numpy as np

def preprocess_image(image_np):
    gray = cv2.cvtColor(image_np, cv2.COLOR_BGR2GRAY)
    # Resize the image
    image_resized = cv2.resize(gray, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)
    # Denoise the image
    denoised = cv2.GaussianBlur(image_resized, (5, 5), 0)
    # Apply adaptive thresholding
    processed_image = cv2.adaptiveThreshold(
        denoised, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2
    )
    return processed_image

def extract_text_from_image(image):
    # Convert PIL image to NumPy array
    image_np = np.array(image)
    # Preprocess the image
    processed_image = preprocess_image(image_np)
    # Convert back to PIL format for Tesseract
    pil_image = Image.fromarray(processed_image)
    # Extract text using Tesseract
    extracted_text = pytesseract.image_to_string(pil_image, lang="tel")
    return extracted_text


# Build the Gradio interface
interface = gr.Interface(
    fn=extract_text_from_image,           # Function to process input
    inputs=gr.Image(type="pil"),          # Input: Upload an image
    outputs="text",                       # Output: Extracted text
    title="Telugu Text Extractor",
    description="Upload an image containing text in Telugu. This tool will extract and display the text."
)

# Launch the interface
interface.launch()
```
